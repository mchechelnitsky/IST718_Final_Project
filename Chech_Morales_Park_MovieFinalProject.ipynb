```python
import pandas as pd
```

```python
#base data set
base_data = pd.read_csv("base_data.csv")
#cleaning step 1: drop unnecessary columns
base_data = base_data.drop(['belongs_to_collection','homepage','poster_path','original_title','crew'], axis=1)
base_data = base_data.set_index('id')
#shuffle title to lead column
names = base_data['title']
base_data = base_data.drop(['title'],axis=1)
base_data.insert(0,'Title',names)
base_data['Title'] = base_data['Title'].str.lower()
base_data.shape
```

```python
imdb_df = pd.read_csv("imdb.csv")
imdb_df = imdb_df.drop(['Unnamed: 44','Unnamed: 45','Unnamed: 46','Unnamed: 47'],axis=1)
#imdb_df.head()
```

```python
data = base_data.join(imdb_df.set_index('wordsInTitle'),on='Title',how='inner')
data = data.dropna()
print(data.shape)
data.head()
```
